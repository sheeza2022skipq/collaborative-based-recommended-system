# Popularity based Recommender system

```python
Books_name_ratings = ratings.merge(books,on='ISBN')
```

```python
Books_name_ratings.shape
```

```python
Books_name_ratings
```

```python
numbers_of_rating = Books_name_ratings.groupby('Book-Title').count()['Book-Rating'].reset_index()
numbers_of_rating.rename(columns={'Book-Rating':'Num-Rating'},inplace =True)
```

```python
numbers_of_rating
```

```python
avg_of_rating = Books_name_ratings.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_of_rating.rename(columns={'Book-Rating':'Avg-Rating'},inplace =True)
```

```python
avg_of_rating
```

```python
popularity = numbers_of_rating.merge(avg_of_rating,on='Book-Title')
popularity
```

```python
Top_popular = popularity[popularity['Num-Rating']>=250].sort_values('Avg-Rating',ascending = False)
Top_popular.head()
```

```python
top_fifty = Top_popular.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title', 'Book-Author', 'Num-Rating', 'Avg-Rating', 'Image-URL-L']]
top_fifty.shape
```

```python
top_fifty
```

```python
top_fifty.iloc[0]['Image-URL-L']
```

# Collaborative Based filtering

```python
A = Books_name_ratings.groupby('User-ID').count()['Book-Rating']>200
active_users = A[A].index
```

```python
active_users = A[A].index
```

```python
Filtered_rating = Books_name_ratings[Books_name_ratings['User-ID'].isin(active_users)]
```

```python
Filtered_rating
```

```python
B = Filtered_rating.groupby('Book-Title').count()['Book-Rating'] >= 50
famous_books = B[B].index
```

```python
famous_books
```

```python
Final_rating = Filtered_rating[Filtered_rating['Book-Title'].isin(famous_books)]
```

```python
Final_rating
```

```python
pivottable = Final_rating.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pivottable
```

```python
pivottable.fillna(0,inplace=True)
```

```python
pivottable
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pivottable)
```

```python
similarity_score.shape
```

# Recommender

```python
def recommender(book_name):
    #fatch index
    index = np.where(pivottable.index == book_name)[0][0]
    recommended_items = sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:10]
    for i in recommended_items:
        print(pivottable.index[i[0]])
```

```python
recommender('1984')
```
